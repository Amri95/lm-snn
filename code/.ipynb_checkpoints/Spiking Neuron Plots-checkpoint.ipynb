```python
from brian2 import *

%matplotlib inline
```

```python
start_scope()

eqs = '''
dv/dt = (I - v) / tau : 1
I : 1
tau : second
'''

P = PoissonGroup(1, np.arange(1) * Hz + 10*Hz)
G = NeuronGroup(1, eqs, threshold='v > 1', reset='v = 0')
G.I = 2
G.tau = 10 * ms

M = StateMonitor(G, 'v', record=True)
spikemon = SpikeMonitor(G)

run(100 * ms)

plot(M.t / ms, M.v[0], '-b', label='Neuron 0')
for t in spikemon.t:
    axvline(t/ms, ls='--', c='r', lw=3)
xlabel('Time (ms)')
ylabel('v')
legend(loc='best')
```
