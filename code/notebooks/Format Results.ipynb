```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# modify this to suit your directory structure if needed
top_level_path = '../../'

# plotting inline of notebook magic function
%matplotlib inline

# setting size of inline figures
mpl.rcParams['figure.figsize'] = (12, 10)

results = pd.read_csv(top_level_path + 'data/all_accuracy_results.csv', delimiter=',', encoding="utf-8-sig", header=0)
results['Network'] = pd.Series([ results['Network'].loc[idx].replace('_no_weight_dependence', '') \
                                                                for idx in xrange(len(results)) ])
results['Network'] = pd.Series([ results['Network'].loc[idx].replace('_weight_dependence', '') \
                                                                for idx in xrange(len(results)) ])
results['Network'] = pd.Series([ results['Network'].loc[idx].replace('_no_postpre', '') \
                                                                for idx in xrange(len(results)) ])
results['Network'] = pd.Series([ results['Network'].loc[idx].replace('_postpre', '') \
                                                                for idx in xrange(len(results)) ])
results['Network'] = pd.Series([ results['Network'].loc[idx].replace('no_weight_sharing', 'NWS') \
                                                                for idx in xrange(len(results)) ])
results['Network'] = pd.Series([ results['Network'].loc[idx].replace('weight_sharing', 'WS') \
                                                                for idx in xrange(len(results)) ])
results['Network'] = pd.Series([ ' '.join(results['Network'].loc[idx].split('_')[2:-1]) \
                                                                for idx in xrange(len(results)) ])

# results.sort_values(['All'], ascending=False)
```

```python
results['Connectivity'] = pd.Series([ results['Network'].loc[idx].split(' ')[0] for idx in xrange(len(results)) ])
results['Conv. Size'] = pd.Series([ int(results['Network'].loc[idx].split(' ')[1]) for idx in xrange(len(results)) ])
results['Stride'] = pd.Series([ int(results['Network'].loc[idx].split(' ')[2]) for idx in xrange(len(results)) ])
results['No. Patches'] = pd.Series([ int(results['Network'].loc[idx].split(' ')[3]) for idx in xrange(len(results)) ])
results['Weight Sharing'] = pd.Series([ results['Network'].loc[idx].split(' ')[5] for idx in xrange(len(results)) ])
results['Lattice Structure'] = pd.Series([ int(results['Network'].loc[idx].split(' ')[6]) for idx in xrange(len(results)) ])

results = results.drop('Network', 1)

results = results.sort_values(['Connectivity', 'Conv. Size'])
results = results.set_index(['Connectivity', 'Conv. Size'])
```

```python
sub_tables = {}
for weight_sharing in ['WS', 'NWS']:
    for lattice_structure in [8]:
        for n_patches in [50]:
            sub_table = results.loc[(results['Weight Sharing'] == weight_sharing) & \
                (results['Lattice Structure'] == lattice_structure) & \
                (results['No. Patches'] == n_patches) & \
                (results['Stride'] == 1) | (results['Stride'] == 2) ].drop(['Lattice Structure', \
                'Weight Sharing', 'No. Patches', 'Stride'], 1).to_latex(index=True, \
                columns=['All', 'Most-spiked', 'Top-percent', 'Corr. Clustering'])
            sub_tables[(lattice_structure, weight_sharing, n_patches)] = \
                            ((lattice_structure, weight_sharing, n_patches), sub_table)
```
